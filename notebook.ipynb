### Spliting data into train and validation set

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(df,test_size=0.2,stratify=df['VAR21'])
```

```python
print(train['VAR21'].value_counts()/train.shape[0])
print(val['VAR21'].value_counts()/val.shape[0])
```

```python
Xtrain = train.drop(columns='VAR21')
ytrain = train['VAR21']
Xval = val.drop(columns='VAR21')
yval = val['VAR21']
```

### Imputing Missing Values

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
```

```python
ii = IterativeImputer(max_iter=100,initial_strategy='median',min_value=0,max_value=1000,random_state=420 )
knn = KNNImputer(n_neighbors=3) 
```

```python
ii.fit(Xtrain)
knn.fit(Xtrain)
Xtrain_ii = pd.DataFrame(ii.transform(Xtrain),index=Xtrain.index,columns=Xtrain.columns)
Xtrain_knn = pd.DataFrame(knn.transform(Xtrain),index=Xtrain.index,columns=Xtrain.columns)
```

```python
ytrain = ytrain.replace({'Low':0,'Medium':1,'High':2})
```

```python
ixtrain,ixtest,iytrain, iytest = train_test_split(Xtrain_ii,ytrain,test_size=0.2,stratify=ytrain,random_state=42)
kxtrain,kxtest,kytrain,kytest = train_test_split(Xtrain_knn,ytrain,test_size=0.2,stratify=ytrain,random_state=42)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
etc = ExtraTreesClassifier()
etc.fit(ixtrain,iytrain)
print(etc.score(ixtest,iytest))
etc.fit(kxtrain,kytrain)
print(etc.score(kxtest,kytest))
```

###### We are going to use Iterative Imputer since it gives high accuracy than KNN.

```python
yval
```

```python
# Using Iterative Imputer on Validation set
Xval =  pd.DataFrame(ii.transform(Xval),index=Xval.index,columns=Xval.columns)
yval = yval.replace({'Low':0,'Medium':1,'High':2})
```

```python
Xval.isna().sum()
```

### Feature Scaling

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import RobustScaler
PT = PowerTransformer()
RS = RobustScaler()
```

```python
xtrain_rs = RS.fit_transform(Xtrain_ii)
Xval_rs = RS.transform(Xval)
```

```python
xtrain_pt = PT.fit_transform(Xtrain_ii)
Xval_pt = PT.transform(Xval)
```

#### Using SMOTE for Oversampling

```python
#Using SMOTE
from imblearn.over_sampling import SMOTE
smt = SMOTE(random_state=51)
```

```python
xtrain_smtp, ytrain_smtp = smt.fit_resample(xtrain_pt,ytrain) #SMOTE+ Power Transformer
xtrain_smtr, ytrain_smtr = smt.fit_resample(xtrain_rs,ytrain) # SMOTE + Robust Scaler
```

```python
# importing necessary+ models
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, IsolationForest
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
svm = SVC()
xgb = XGBClassifier(use_label_encoder=False, eval_metric='merror')
rf = RandomForestClassifier()
isf = IsolationForest()
```

```python
def result(clf,xtrain,ytrain,xval,yval):
    clf.fit(xtrain,ytrain)
    pred = clf.predict(xtrain)
    print("f1_score for train set is:", f1_score(ytrain,pred,average='micro'))
    pred = clf.predict(xval)
    print("f1_score for validation set is:", f1_score(yval,pred,average='micro'))
```

### Using XGBClassifier 

```python
#for Robust Scaler
result(xgb,xtrain_rs,ytrain,Xval_rs,yval)
```

```python
#for Robust Scaler + Smote
result(xgb,xtrain_smtr,ytrain_smtr,Xval_rs,yval)
```

```python
#for Power Transformer
result(xgb,xtrain_pt,ytrain,Xval_pt,yval)
```

```python
#for Power Transformer + Smote
result(xgb,xtrain_smtp,ytrain_smtp,Xval_pt,yval)
```

##### Robust Scaler is working the best on validation set.

```python
result(svm,xtrain_rs,ytrain,Xval_rs,yval)
```

```python
result(rf,xtrain_rs,ytrain,Xval_rs,yval)
```

```python
result(isf,xtrain_rs,ytrain,Xval_rs,yval)
```

##### Robust Scaler with XGBClassifier is performing the best here.

### Hyperparameter Tuning

#### For XGBClassifier

```python
from sklearn.model_selection import GridSearchCV

params = {
        'learning_rate': [0.03, 0.06, 0.1],
        'min_child_weight': [1, 5, 2],
        'gamma': [0,1,3],
        'subsample': [0.5,1],
        'max_depth': [3,6]
        }

gridsearch = GridSearchCV(xgb, param_grid=params, scoring='accuracy', n_jobs=1, cv=3, verbose=2)
```

```python
gridsearch.fit(xtrain_rs, ytrain)
```

```python
gridsearch.best_params_
```

```python
def Final_result(clf,xtrain,ytrain,xval,yval):
    pred = clf.predict(xtrain)
    print("f1_score for train set is:", f1_score(ytrain,pred,average='micro'))
    pred = clf.predict(xval)
    print("f1_score for validation set is:", f1_score(yval,pred,average='micro'))
```

```python
Final_result(gridsearch,xtrain_rs,ytrain,Xval_rs,yval)
```

#### For RandomForestClassifier

```python
rf_params = {'n_estimators': [100, 200, 300],
                'max_depth': [5, 10, None],
                'min_samples_split' : [2,5],
                'max_features': [0.5, 'auto']
}
gridsearch_rf = GridSearchCV(rf, param_grid=rf_params, scoring='accuracy', n_jobs=1, cv=3, verbose=2)
```

```python
gridsearch_rf.fit(xtrain_rs, ytrain)
```

```python
gridsearch_rf.best_params_
```

```python
Final_result(gridsearch_rf,xtrain_rs,ytrain,Xval_rs,yval)
```
